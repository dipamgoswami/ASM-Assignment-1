```python
import pandas as pd
```

```python
df = pd.read_csv("covid.csv")
df.head()
```

```python
df= df[['EL', 'ER' , 'SL', 'SR']]
df.head()
```

```python
type(df['EL'][1])
```

```python
from datetime import datetime
df['EL'] = pd.to_datetime(df['EL'])
df['ER'] = pd.to_datetime(df['ER'])
df['SL'] = pd.to_datetime(df['SL'])
df['SR'] = pd.to_datetime(df['SR'])
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.fillna('2019-12-01 00:00:00')
df.head()
```

```python
for i in range(len(df)):
    k1 = pd.Timestamp(df['EL'][i]) - pd.Timestamp('2019-12-01 00:00:00')
    df['EL'][i] = k1.days + (k1.seconds)/86400
    
    k2 = pd.Timestamp(df['ER'][i]) - pd.Timestamp('2019-12-01 00:00:00')
    df['ER'][i] = k2.days + (k2.seconds)/86400
    
    k3 = pd.Timestamp(df['SL'][i]) - pd.Timestamp('2019-12-01 00:00:00')
    df['SL'][i] = k3.days + (k3.seconds)/86400
    
    k4 = pd.Timestamp(df['SR'][i]) - pd.Timestamp('2019-12-01 00:00:00')
    df['SR'][i] = k4.days + (k4.seconds)/86400
```
